```python

import sys #library for termination of the code

name = input("Enter name: ")
print("<><><><><><><><><><><><><><><><><>") #SPACE

#PRELIM GRADE

#to get the prelim grade, we need to get the ff scores:
print("ENTER THE PRELIM SCORES: ")
PHOA = float(input("Enter prelim hands-on activity grade (100 ITEMS): "))
if PHOA > 100 < 0:
    print("Invalid Input! PLEASE RESTART THE CODE")
    sys.exit() #EXIT THE CODE IF ITS TRUE

PQUIZ = float(input("Enter prelim quiz grade (100 ITEMS): "))
if PQUIZ > 100 < 0:
    print("Invalid Input! PLEASE RESTART THE CODE")
    sys.exit()

PAS = float(input("Enter prelim assignment grade (100 ITEMS): "))
if PAS > 100 < 0:
    print("Invalid Input! PLEASE RESTART THE CODE")
    sys.exit()

print("----------------------") #SPACE
#Calculate Prelim Class Standing (CS):
PCS = PHOA*.5 + PQUIZ*.3 + PAS*.2
print("Prelim CS is: ", PCS)
print("----------------------") #SPACE
#Prelim Exam:
PEX = float(input("Enter prelim exam score: "))
#Compute prelim Grade
PG = PEX*.5 + PCS*.5
print("Prelim Grade is: ", PG)
print("----------------------") #SPACE

#MIDTERM GRADE

#To compute for midterm grade, we need to get the ff scores:
print("ENTER THE MIDTERM SCORES: ")
MHOA = float(input("Enter midterm hands-on activity grade (100 ITEMS): "))
if MHOA > 100 < 0:
    print("Invalid Input! PLEASE RESTART THE CODE")
    sys.exit()

MQUIZ = float(input("Enter midterm quiz grade (100 ITEMS): "))
if MQUIZ > 100 < 0:
    print("Invalid Input! PLEASE RESTART THE CODE")
    sys.exit()

MAS = float(input("Enter midterm assignment grade (100 ITEMS): "))
if MAS > 100 < 0:
    print("Invalid Input! PLEASE RESTART THE CODE")
    sys.exit()

print("----------------------") #SPACE
#Calculate Midterm Class Standing (CS):
MCS = MHOA*.5 + MQUIZ*.3 + MAS*.2
print("Midterm CS is: ", MCS)
print("----------------------") #SPACE
#Midterm Exam:
MEX = float(input("Enter midterm exam score: "))
#Compute midterm Grade
MG = (PG*1/3) + (2/3)*(MCS*.5 + MEX*.5)
print("Midterm Grade is: ", MG)
print("----------------------") #SPACE

#FINAL GRADE
#To compute for final grade, we need to get the ff scores:
print("ENTER THE FINAL SCORES: ")
FHOA = float(input("Enter final hands-on activity grade (100 ITEMS): "))
if FHOA > 100 < 0:
    print("Invalid Input! PLEASE RESTART THE CODE")
    sys.exit()

FQUIZ = float(input("Enter final quiz grade (100 ITEMS): "))
if FQUIZ > 100 < 0:
    print("Invalid Input! PLEASE RESTART THE CODE")
    sys.exit()

FAS = float(input("Enter final assignment grade (100 ITEMS): "))
if FAS > 100 < 0:
    print("Invalid Input! PLEASE RESTART THE CODE")
    sys.exit()

print("----------------------") #SPACE
#Calculate Final Class Standing (CS):
FCS = FHOA*.5 + FQUIZ*.3 + FAS*.2
print("Final CS is: ", FCS)
print("----------------------") #SPACE
#Final exam:
FEX = float(input("Enter final exam score: "))
#Compute final Grade:
FG = (MG*1/3) + (2/3)*(FCS*.5 + FEX*.5)
print("Final Grade is: ", FG)
100

#SUMMARY OF THE STUDENT GRADES:
print("~~~~~~~~~~~~~~~~~~~~~~~~~~") #SPACE
print("SUMMARY OF THE STUDENT GRADES")
print("Name: ",name)
print("PRELIM GRADE: ", PG)
print("MIDTERM GRADE: ", MG)
print("FINAL GRADE: ", FG)
print("~~~~~~~~~~~~~~~~~~~~~~~~~~") #SPACE

if FG >= 99 < 100:
    print("FINAL GRADE POINT EQUIVALENCE: 1.00")
elif FG >= 95 < 98:
    print("FINAL GRADE POINT EQUIVALENCE: 1.25")
elif FG >= 90 < 94:
    print("FINAL GRADE POINT EQUIVALENCE: 1.50")
elif FG >= 85 < 89:
    print("FINAL GRADE POINT EQUIVALENCE: 1.75")
elif FG >= 80 < 84:
    print("FINAL GRADE POINT EQUIVALENCE: 2.00")
elif FG >= 75 < 79:
    print("FINAL GRADE POINT EQUIVALENCE: 2.25")
elif FG >= 70 < 74:
    print("FINAL GRADE POINT EQUIVALENCE: 2.50")
elif FG >= 65 < 69:
    print("FINAL GRADE POINT EQUIVALENCE: 2.75")
elif FG >= 60 < 64:
    print("FINAL GRADE POINT EQUIVALENCE: 3.00")
else:
    print("FINAL GRADE POINT EQUIVALENCE: 5.00")
```
