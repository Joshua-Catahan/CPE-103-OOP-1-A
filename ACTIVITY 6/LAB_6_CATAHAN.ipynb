# **Creating the Classes**

Character.py

```python
class Character:
    def __init__(self, username):
        self.__username = username
        self.__hp = 100
        self.__mana = 100
        self.__damage = 5
        self.__str = 0  # strength stat
        self.__vit = 0  # vitality stat
        self.__int = 0  # intelligence stat
        self.__agi = 0  # agility stat

    def getUsername(self):
        return self.__username

    def setUsername(self, new_username):
        self.__username = new_username

    def getHp(self):
        return self.__hp

    def setHp(self, new_hp):
        self.__hp = new_hp

    def getDamage(self):
        return self.__damage

    def setDamage(self, new_damage):
        self.__damage = new_damage

    def getStr(self):
        return self.__str

    def setStr(self, new_str):
        self.__str = new_str

    def getVit(self):
        return self.__vit

    def setVit(self, new_vit):
        self.__vit = new_vit

    def getInt(self):
        return self.__int

    def setInt(self, new_int):
        self.__int = new_int

    def getAgi(self):
        return self.__agi

    def setAgi(self, new_agi):
        self.__agi = new_agi

    def reduceHp(self, damage_amount):
        self.__hp = self.__hp - damage_amount

    def addHp(self, heal_amount):
        self.__hp = self.__hp + heal_amount
```

# **Multi-level Inheritance**

Swordsman.py

```python
class Swordsman (Novice):
    def __init__(self, username):
        super().__init__(username)
        self.setStr(5)
        self.setVit(10)
        self.setHp(self.getHp()+self.getVit())

    def slashAttack(self, character):
        self.new_damage = self.getDamage()+self.getStr()
        character.reduceHp(self.new_damage)
        print(f"{self.getUsername()} performed Slash Attack! -{self.new_damage}")
```

Archer.py

```python
class Archer (Novice):
    def __init__(self, username):
        super().__init__(username)
        self.setAgi(5)
        self.setInt(5)
        self.setVit(5)
        self.setHp(self.getHp()+self.getVit())

    def rangedAttack(self, character):
        import random
        self.new_damage = self.getDamage()+random.randint(0,self.getInt())
        character.reduceHp(self.new_damage)
        print(f"{self.getUsername()} performed Slash Attack! -{self.new_damage}")
```

Magician.py

```python
class Magician (Novice):
    def __init__(self, username):
        super().__init__(username)
        self.setInt(10)
        self.setVit(5)
        self.setHp(self.getHp()+self.getVit())

    def heal(self):
        self.addHp(self.getInt())
        print(f"{self.getUsername()} performed Heal! +{self.getInt()}")

    def magicAttack(self, character):
        self.new_damage = self.getDamage()+self.getInt()
        character.reduceHp(self.new_damage)
        print(f"{self.getUsername()} performed Magic Attack! -{self.new_damage}")
```

# **Task:**

Create a new file Game.py inside the same folder use the pre-made classes to create a simple Game where two players or
one player vs a computer will be able to reduce their opponent’s hp to 0.

Requirements:
1.	The game must be able to select between 2 modes: Single player and Player vs Player. The game can spawn multiple matches where single player or player vs player can take place.
2.	In Single player:
•	the player must start as a Novice, then after 2 wins, the player should be able to select a new role between Swordsman, Archer, and Magician.
•	The opponent will always be a boss named Monster.
3.	In Player vs Player, both players must be able to select among all the possible roles available except Boss.
4.	Turns of each player for both modes should be randomized and the match should end when one of the players hp is zero.
5.	Wins of each player in a game for both the modes should be counted


```python
import random

class Character:
    def __init__(self, username):
        self.__username = username
        self.__hp = 100
        self.__mana = 100
        self.__damage = 5
        self.__str = 0  # Strength stat
        self.__vit = 0  # Vitality stat
        self.__int = 0  # Intelligence stat
        self.__agi = 0  # Agility stat
        self.wins = 0
        self.default_hp = 100  # Store initial HP for resetting

    def getUsername(self):
        return self.__username

    def getHp(self):
        return self.__hp

    def setHp(self, new_hp):
        self.__hp = max(0, new_hp)  # Ensure HP doesn't go negative

    def getDamage(self):
        return self.__damage

    def reduceHp(self, damage_amount):
        self.__hp = max(0, self.__hp - damage_amount)  # Prevent negative HP

    def is_alive(self):
        return self.__hp > 0

    def reset_health(self):
        self.__hp = self.default_hp  # Reset HP to default

    def display_status(self):
        print(f"{self.getUsername()} - HP: {self.getHp()} | Wins: {self.wins}")

    def setStr(self, value): self.__str = value
    def getStr(self): return self.__str

    def setVit(self, value): self.__vit = value
    def getVit(self): return self.__vit

    def setInt(self, value): self.__int = value
    def getInt(self): return self.__int

    def setAgi(self, value): self.__agi = value
    def getAgi(self): return self.__agi


class Novice(Character):
    def attack(self, target):
        target.reduceHp(self.getDamage()) # Reduce the target's HP by the attacker's damage value
        print(f"{self.getUsername()} performed Basic Attack! -{self.getDamage()}")


class Swordsman(Novice):
    def __init__(self, username):
        super().__init__(username)
        self.setStr(5)
        self.setVit(10)
        self.setHp(self.getHp() + self.getVit())

    def attack(self, target):
        choice = input("Choose attack: 1. Basic Attack 2. Slash Attack\n")
        damage = self.getDamage() + (self.getStr() if choice == "2" else 0)  # Calculate damage: add strength bonus if Slash Attack is chosen
        print(f"{self.getUsername()} performed {'Slash Attack' if choice == '2' else 'Basic Attack'}! -{damage}")
        target.reduceHp(damage)
      #if 2 is not chosen, it will automatically use "1. basic attack" which will only use self.getDamage() without added stats bonus.

class Archer(Novice):
    def __init__(self, username):
        super().__init__(username)
        self.setAgi(5)
        self.setInt(5)
        self.setVit(5)
        self.setHp(self.getHp() + self.getVit())

    def attack(self, target):
        choice = input("Choose attack: 1. Basic Attack 2. Ranged Attack\n")
        damage = self.getDamage() + (random.randint(0, self.getInt()) if choice == "2" else 0) # Calculate damage: Ranged Attack adds a random bonus based on intelligence
        print(f"{self.getUsername()} performed {'Ranged Attack' if choice == '2' else 'Basic Attack'}! -{damage}")
        target.reduceHp(damage)
      #if 2 is not chosen, it will automatically use "1. basic attack" which will only use self.getDamage() without added stats bonus.


class Magician(Novice):
    def __init__(self, username):
        super().__init__(username)
        self.setInt(10)
        self.setVit(5)
        self.setHp(self.getHp() + self.getVit())

    def attack(self, target):
        choice = input("Choose attack: 1. Basic Attack 2. Magic Attack 3. Heal\n")
        if choice == "3":
            # Heal the Magician instead of attacking
            heal_amount = self.getInt()
            self.setHp(self.getHp() + heal_amount)
            print(f"{self.getUsername()} performed Heal! +{heal_amount}")
        else:
            damage = self.getDamage() + (self.getInt() if choice == "2" else 0)  # Calculate damage: Magic Attack adds intelligence bonus
            print(f"{self.getUsername()} performed {'Magic Attack' if choice == '2' else 'Basic Attack'}! -{damage}")
            target.reduceHp(damage)
      #if 2 and 3 are not chosen, it will automatically use "1. basic attack" which will only use self.getDamage() without added stats bonus.

class Boss(Character):
    def __init__(self, username):
        super().__init__(username)
        self.setStr(10)
        self.setVit(25)
        self.setInt(5)
        self.setHp(self.getHp() + self.getVit())
        self.default_hp = self.getHp()

    def attack(self, target):
        # Boss randomly chooses an attack or heal action
        choice = random.choice(["1", "2", "3"])
        if choice == "3":
            # Heal itself instead of attacking
            heal_amount = self.getInt()
            self.setHp(self.getHp() + heal_amount)
            print(f"{self.getUsername()} performed Heal! +{heal_amount}")
        else:
            damage = self.getDamage() + (self.getInt() if choice == "2" else 0)  # Calculate damage: Magic Attack adds intelligence bonus
            print(f"{self.getUsername()} attacks {target.getUsername()} with {'Magic Attack' if choice == '2' else 'Basic Attack'}! -{damage}")
            target.reduceHp(damage)
      #if 2 and 3 are not chosen, it will automatically use "1. basic attack" which will only use self.getDamage() without added stats bonus.


def game_mode():
    print("\nChoose a game mode:")
    print("1. Single Player")
    print("2. Player vs Player")
    game_choice = input("Enter the number of your choice: ")

    if game_choice == "1":
        single_player()
    elif game_choice == "2":
        player_vs_player()
    else:
        print("Invalid choice. Please choose again.")

def choose_role(name):
    print("Choose your role:")
    print("1. Swordsman")
    print("2. Archer")
    print("3. Magician")
    player_choice = input("Enter the number of your choice: ")

    roles = {"1": Swordsman, "2": Archer, "3": Magician}
    return roles.get(player_choice, Swordsman)(name)

def ask_new_mode():
    # Ask the player if they want to start a new game mode
    choice = input("Do you want to select a new mode? (y/n): ").lower()
    if choice == "y":
        game_mode()

def single_player():
    # Get player’s name and initialize as Novice
    name = input("Enter your character's name: ")
    player = Novice(name)  # Player starts as a Novice
    monster = Boss("Monster")  # Opponent is always a Boss named "Monster"
    role_changed = False  # Track if the player has changed role

    while True:
        # Start a battle between the player and the monster
        fight(player, monster)

        # Allow the player to change role after 2 wins
        if player.wins >= 2 and not role_changed:
            print("\nYou have won 2 times! You can now choose a new class.")
            new_role = choose_role(player.getUsername())
            new_role.wins = player.wins  # Retain win count
            player = new_role
            role_changed = True  # Ensure role change only happens once

        # Reset player's HP if they lose
        if not player.is_alive():
            player.reset_health()

        # Reset the monster if it dies
        if not monster.is_alive():
            monster_wins = monster.wins  # Preserve the monster's win count
            monster = Boss("Monster")  # Create a new monster
            monster.wins = monster_wins  # Restore win count

        choice = input("Continue fighting? (y/n): ")
        if choice != "y":
            break

    print("\nGame Over!")
    ask_new_mode()

def player_vs_player():
    print("\nPlayer 1, enter your username:")
    player1 = choose_role(input("Enter username: "))
    print("\nPlayer 2, enter your username:")
    player2 = choose_role(input("Enter username: "))

    while True:
        # Start a fight between the two players
        fight(player1, player2)

        # Reset both players' health after the fight
        player1.reset_health()
        player2.reset_health()

        choice = input("Play again? (y/n): ")
        if choice != "y":
            break

    print("\nGame Over!")
    ask_new_mode()

def fight(player1, player2):
    print(f"\nBattle Start: {player1.getUsername()} vs {player2.getUsername()}")

    players = [player1, player2]  # Store players in a list for easier management
    while player1.is_alive() and player2.is_alive():

        # Randomized turns: a player may attack multiple times in a row
        while random.choice([True, False]):
            print(f"{players[0].getUsername()} is attacking!")
            players[0].attack(players[1])  # Player 1 attacks Player 2
            players[0].display_status()
            players[1].display_status()
            if not players[1].is_alive():
                break  # Stop the attack if the opponent is defeated

        # If Player 2 is still alive, they attack
        if players[1].is_alive():
            while random.choice([True, False]):
                print(f"{players[1].getUsername()} is attacking!")
                players[1].attack(players[0])  # Player 2 attacks Player 1
                players[0].display_status()
                players[1].display_status()
                if not players[0].is_alive():
                    break  # Stop the attack if the opponent is defeated

    # Determine the winner and display the result
    winner = player1 if player1.is_alive() else player2
    print(f"\n{winner.getUsername()} wins!")
    winner.wins += 1  # Increase the winner's win count

    # Display final status of both players
    print("\nFinal Status:")
    player1.display_status()
    player2.display_status()

if __name__ == "__main__":
    game_mode()


# NOTES:

# HEALTH WON'T RESET IN SINGLE PLAYER UNTIL IT REACHES ZERO TO MAKE THE POSSIBILITY OF PLAYER WINNING HIGHER. (PVP EXCLUDED)
# THEREFORE, THE PLAYER AND MONSTER FIGHTS CONTINUES WITH THE REMAINING HEALTH TILL ITS HEALTH REACHES ZERO.
# SINCE THERE ARE ONLY 1 TYPE OF ATTACK FOR NOVICE, THE GAME WILL AUTOMATICALLY PLAY UNTIL THE PLAYER REACHES 2 WINS WHERE THEY CAN SELECT NEW ROLE.
# TURNS ARE RANDOMIZED SO PLAYERS/MONSTERS CAN ATTACK CONSECUTIVELY DEPENDING ON THEIR LUCK.


```
